```python
# Define realistic lead time ranges for each category
lead_time_ranges = {
    'WINE': (14, 21),         # 2-3 weeks
    'BEER': (7, 14),          # 1-2 weeks
    'LIQUOR': (10, 20),       # 1.5-3 weeks
    'KEGS': (14, 21),         # 2-3 weeks
    'STR_SUPPLIES': (3, 7),   # 3-7 days
    'REF': (5, 10),           # 5-10 days
    'NON-ALCOHOL': (5, 10),   # 5-10 days
    'DUNNAGE': (3, 7)         # 3-7 days
}

# Function to estimate lead time based on category
def estimate_lead_time(category):
    if category in lead_time_ranges:
        return np.random.randint(lead_time_ranges[category][0], lead_time_ranges[category][1] + 1)
    else:
        return np.nan  # Handle unexpected categories

# Apply the lead time estimation function
leadtime_info['Lead Time (days)'] = leadtime_info['Category'].apply(estimate_lead_time)

leadtime_info
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Step 1: Generate a reference for SKU, Category, and Supplier
# Assuming 'main' and 'leadtime_info' are already loaded as DataFrames
# We will sample from these columns randomly

# Number of rows for orders_info
n_rows = 170000

# Sample SKU, Category, and Supplier from the existing dataframes
skus = main['SKU'].unique()
categories = main['Category'].unique()
suppliers = main['Supplier'].unique()

# Randomly sample these columns to create the orders_info dataframe
orders_info = pd.DataFrame({
    'SKU': np.random.choice(skus, size=n_rows),
    'Category': np.random.choice(categories, size=n_rows),
    'Supplier': np.random.choice(suppliers, size=n_rows),
})

# Step 2: Estimate Order Quantity (for example, randomly between 1 and 1000)
orders_info['Order Quantity'] = np.random.randint(1, 1001, size=n_rows)

# Step 3: Generate Order Placed Date (randomly within the last 365 days)
start_date = datetime.now() - timedelta(days=365)
orders_info['Order Placed Date'] = [start_date + timedelta(days=np.random.randint(0, 366)) for _ in range(n_rows)]

# Step 4: Generate Availability as a boolean (randomly either True or False)
orders_info['Availability'] = np.random.choice([True, False], size=n_rows, p=[0.8, 0.2])  # 80% chance for True (availability)

# Display the first few rows
orders_info.head()
```

```python
import pandas as pd
import numpy as np

# Assuming orders_info and suppliers are already defined

# Step 1: Generate the availability rates for each supplier using a Beta distribution
n_suppliers = len(suppliers)

# Use Beta distribution to generate availability rates, skewed towards higher values (80%-100%)
availability_rates = np.random.beta(a=5, b=2, size=n_suppliers)  # Skewed towards 1

# Normalize availability so that it's between 0 and 1
availability_rates = availability_rates.clip(0, 1)

# Step 2: Scale and shift availability rates to the 80%-100% range
availability_rates = 0.2 * availability_rates + 0.8  # This will map [0, 1] to [0.8, 1]

# Map the availability rates back to suppliers
supplier_availability = dict(zip(suppliers, availability_rates))

# Step 3: Map the availability rate to the 'Availability' column in orders_info
orders_info['Availability Rate'] = orders_info['Supplier'].map(supplier_availability)

# Step 4: Generate Availability as a boolean (True or False)
# For each order, the availability is True if the random number is less than or equal to the supplier's availability rate
orders_info['Availability'] = np.random.rand(len(orders_info)) <= orders_info['Availability Rate']

# Display the first few rows to check
print(orders_info.head())
```

```python
# Calculate supplier availability rate
supplier_availability_rate = orders_info.groupby('Supplier')['Availability'].mean().reset_index()

# Convert to percentage (multiplying by 100)
supplier_availability_rate['Availability Rate (%)'] = supplier_availability_rate['Availability'] * 100

# Display the availability rates for each supplier
supplier_availability_rate
```

```python
import pandas as pd
import numpy as np

# Assuming orders_info and leadtime_info are already defined

# Step 1: Ensure 'Order Placed Date' is in datetime format
orders_info['Order Placed Date'] = pd.to_datetime(orders_info['Order Placed Date'], errors='coerce')

# Step 2: Merge orders_info with leadtime_info to get the expected lead time for each SKU
orders_info = orders_info.merge(leadtime_info[['SKU', 'Lead Time (days)']], on='SKU', how='left')

# Step 3: Estimate the delivery date for orders with availability as True
np.random.seed(42)  # For reproducibility

def estimate_delivery_date(row):
    if row['Availability']:  # For orders with availability as True
        # Ensure the lead time is not NaN (if no match was found in leadtime_info)
        if not pd.isna(row['Lead Time (days)']):
            # Add a random variability (between -5 to +5 days) to the lead time
            lead_time_variability = np.random.randint(-5, 6)
            estimated_lead_time = row['Lead Time (days)'] + lead_time_variability
            
            # Calculate the estimated delivery date by adding lead time to the order date
            order_date = row['Order Placed Date']
            estimated_delivery_date = order_date + pd.Timedelta(days=estimated_lead_time)
            return estimated_delivery_date
        else:
            return pd.NaT  # If lead time is missing, return NA
    else:
        return pd.NaT  # For unavailable orders, set as NA

# Apply the function to estimate delivery dates
orders_info['Estimated Delivery Date'] = orders_info.apply(estimate_delivery_date, axis=1)

# Step 4: Compare the estimated delivery date with the expected delivery date and flag the delivery status
def flag_delivery_status(row):
    if pd.isna(row['Estimated Delivery Date']):
        return pd.NA  # For unavailable orders, return NA
    else:
        expected_delivery_date = row['Order Placed Date'] + pd.Timedelta(days=row['Lead Time (days)'])
        
        # Flag the delivery status based on the comparison
        if row['Estimated Delivery Date'] < expected_delivery_date:
            return 'early'
        elif row['Estimated Delivery Date'] == expected_delivery_date:
            return 'on-time'
        else:
            return 'late'

# Apply the function to flag the delivery status
orders_info['Delivery Status'] = orders_info.apply(flag_delivery_status, axis=1)

# Display the updated dataframe
print(orders_info[['SKU', 'Availability', 'Order Placed Date', 'Estimated Delivery Date', 'Delivery Status']].head())
```

```python
import numpy as np

# Set the percentage of accurate orders and the maximum deviation
accuracy_rate = 0.95  # 95% of orders are accurate
deviation_rate = 0.05  # 5% of orders have a deviation
max_deviation = 5  # Maximum deviation in quantity, e.g., ±2 units

# Simulate 'Quantity Received' column with small deviations, most are accurate
orders_info['Quantity Received'] = orders_info['Order Quantity']  # Start with accurate quantities

# Introduce small deviations for a small percentage of orders
deviations = np.random.choice([0, np.random.randint(1, max_deviation + 1), -np.random.randint(1, max_deviation + 1)],
                              size=len(orders_info), p=[accuracy_rate, deviation_rate / 2, deviation_rate / 2])

# Apply the deviations to the 'Quantity Received'
orders_info['Quantity Received'] = orders_info['Quantity Received'] + deviations

# Ensure the quantity received is always a whole number (integer)
orders_info['Quantity Received'] = np.round(orders_info['Quantity Received']).astype(int)

# Create 'Quantity Accuracy' flag: 1 if Quantity Received matches Order Quantity, 0 otherwise
orders_info['Quantity Accuracy'] = np.where(orders_info['Order Quantity'] == orders_info['Quantity Received'], 1, 0)

orders_info
```

```python
# Calculate the overall Quantity Accuracy Rate
quantity_accuracy_rate = orders_info['Quantity Accuracy'].mean()

# Print the result as a percentage
print(f"Quantity Accuracy Rate: {quantity_accuracy_rate * 100:.2f}%")
```
