```python
from fuzzywuzzy import fuzz

def standardize_descriptions(df):
    for item_code, group in df.groupby('ITEM CODE'):
        descriptions = group['ITEM DESCRIPTION'].tolist()
        
        # Find similar descriptions using fuzzy matching
        for i in range(len(descriptions)):
            for j in range(i+1, len(descriptions)):
                similarity = fuzz.ratio(descriptions[i], descriptions[j])
                if similarity > 90:  # Adjust threshold as needed
                    # Choose the most common or the first description
                    most_common = max(set(descriptions), key=descriptions.count)
                    descriptions[i] = most_common
                    descriptions[j] = most_common

        # Update the DataFrame
        df.loc[group.index, 'ITEM DESCRIPTION'] = descriptions

    return df
```

```python
# Iterate over each row to adjust the unit_cost
adjusted_unit_cost = []

for index, row in item_info.iterrows():
    total_sales = int(row['TOTAL SALES'])
    unit_cost = row['Unit Cost']
    
    if total_sales <= 0:
        adjusted_unit_cost.append(unit_cost)  # Handle edge case where total_sales is 0 or negative
        continue
    
    # Find all factors of total_sales
    factors = [i for i in range(1, total_sales + 1) if total_sales % i == 0]
    
    # If no factors are found (shouldn't happen for positive total_sales), skip this row
    if not factors:
        adjusted_unit_cost.append(unit_cost)
        continue
    
    # Find the factor closest to unit_cost
    closest_factor = min(factors, key=lambda x: abs(x - unit_cost))
    
    adjusted_unit_cost.append(closest_factor)
    
adjusted_unit_cost
```

```python
# Step 1: Create inventory_info DataFrame with 12 months per ITEM CODE
months = list(range(1, 13))  # 1 to 12
inventory_info = (
    df_2019_cleaned[['ITEM CODE', 'ITEM TYPE']]
    .drop_duplicates()
    .assign(key=1)
    .merge(pd.DataFrame({'MONTH': months, 'key': 1}), on='key')
    .drop('key', axis=1)
)

# Step 2: Merge TOTAL SALES into inventory_info
inventory_info = inventory_info.merge(
    df_2019_cleaned[['ITEM CODE', 'MONTH', 'TOTAL SALES']],
    on=['ITEM CODE', 'MONTH'],
    how='left'
)

inventory_info.head(50)
```

```python
# Step 3: Generate realistic inventory values
def generate_inventory(row):
    total_sales = row['TOTAL SALES']
    
    # Determine stockout
    stockout = np.random.choice([True, False], p=[0.3, 0.7])  # 30% chance of stockout
    if stockout:
        beginning_inventory = total_sales
        ending_inventory = 0
    else:
        beginning_inventory = total_sales * np.random.uniform(1.1, 1.5)  # 10-50% buffer
        ending_inventory = beginning_inventory * np.random.uniform(0.1, 0.5)  # Leftover inventory

    return pd.Series({
        'BEGINNING INVENTORY': round(beginning_inventory, 2),
        'ENDING INVENTORY': round(ending_inventory, 2),
        'STOCKOUT FLAG': stockout
    })

# Apply function to each row
inventory_info[['BEGINNING INVENTORY', 'ENDING INVENTORY', 'STOCKOUT FLAG']] = inventory_info.apply(generate_inventory, axis=1)

# Fill any missing TOTAL SALES with 0 (if there are months with no sales data)
inventory_info['TOTAL SALES'] = inventory_info['TOTAL SALES'].fillna(0)

# Final cleanup
inventory_info = inventory_info[['ITEM CODE', 'ITEM TYPE', 'MONTH', 'TOTAL SALES', 'BEGINNING INVENTORY', 'ENDING INVENTORY', 'STOCKOUT FLAG']]

# Check the resulting DataFrame
```

```python
# Ensure inventory_info is sorted by ITEM CODE and MONTH
inventory_info = inventory_info.sort_values(by=['ITEM CODE', 'MONTH']).reset_index(drop=True)

# Iterate through the DataFrame to validate and adjust BEGINNING INVENTORY
for idx in range(1, len(inventory_info)):
    # Get current and previous row
    current_row = inventory_info.iloc[idx]
    previous_row = inventory_info.iloc[idx - 1]
    
    # Ensure we are comparing the same ITEM CODE
    if current_row['ITEM CODE'] == previous_row['ITEM CODE']:
        # Check if BEGINNING INVENTORY is less than the ENDING INVENTORY of the previous month
        if current_row['BEGINNING INVENTORY'] < previous_row['ENDING INVENTORY']:
            # Adjust BEGINNING INVENTORY to match ENDING INVENTORY or simulate restocking
            inventory_info.at[idx, 'BEGINNING INVENTORY'] = previous_row['ENDING INVENTORY']
        
        # Special case: No sales in the current month
        if current_row['TOTAL SALES'] == 0:
            inventory_info.at[idx, 'BEGINNING INVENTORY'] = max(
                inventory_info.at[idx, 'BEGINNING INVENTORY'], previous_row['ENDING INVENTORY']
            )

# Recalculate AVERAGE INVENTORY for all rows
inventory_info['AVERAGE INVENTORY'] = (
    inventory_info['BEGINNING INVENTORY'] + inventory_info['ENDING INVENTORY']
) / 2

# Validate and display the updated DataFrame
inventory_info.head(60)
```
